## Configuração inicial

Este notebook utiliza:
- Ollama local
- modelo `divea-biostats:latest`
- API REST local

Endpoint:
`http://192.168.1.154:11434`


```
import requests
import json
import pandas as pd
from pathlib import Path
```

## Função de consulta ao Ollama

```
OLLAMA_URL = "http://192.168.1.154:11434/api/generate"
MODEL_NAME = "divea-biostats:latest"

def perguntar_ollama(prompt,
                      model=MODEL_NAME,
                      temperature=0.2):
    
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False,
        "options": {
            "temperature": temperature
        }
    }

    response = requests.post(
        OLLAMA_URL,
        json=payload
    )

    response.raise_for_status()

    resultado = response.json()

    return resultado["response"]
```

## Teste básico do modelo

```
prompt = '''
Explique regressão de Poisson aplicada à epidemiologia,
incluindo interpretação da razão de incidência.
'''

resposta = perguntar_ollama(prompt)

print(resposta)
```

## Auxílio metodológico para o doutorado

```
metodologia = '''
O estudo utiliza séries temporais epidemiológicas
de SRAG provenientes do SIVEP-Gripe.
Serão aplicados modelos LSTM e Temporal Fusion Transformer.
'''

prompt = f'''
Revise metodologicamente o seguinte texto e sugira melhorias
estatísticas e epidemiológicas:

{metodologia}
'''

resposta = perguntar_ollama(prompt)

print(resposta)
```

## Interpretação automática de resultados

```
resultado_modelo = '''
RMSE = 0.42
MAE = 0.31
R² = 0.81
'''

prompt = f'''
Interprete os seguintes resultados de um modelo epidemiológico:

{resultado_modelo}

Explique:
- qualidade preditiva
- possíveis limitações
- implicações epidemiológicas
'''

resposta = perguntar_ollama(prompt)

print(resposta)
```

## Resumo de artigo científico

```
texto_artigo = '''
A interoperabilidade semântica em saúde é fundamental
para integração de sistemas clínicos distribuídos.
O padrão HL7 FHIR tem sido amplamente utilizado.
'''

prompt = f'''
Resuma o texto abaixo em linguagem científica:

{texto_artigo}
'''

resposta = perguntar_ollama(prompt)

print(resposta)
```
