# C_interpretacao_modelos_llm

Interpretação automática de modelos epidemiológicos utilizando LLM local.

## Objetivos
- interpretar métricas
- gerar texto científico
- apoiar discussão epidemiológica

```python
import requests
```

```python
OLLAMA_URL = "http://192.168.1.154:11434/api/generate"
MODEL = "divea-biostats:latest"

def interpretar(prompt):
    payload = {
        "model": MODEL,
        "prompt": prompt,
        "stream": False
    }

    r = requests.post(OLLAMA_URL, json=payload)
    r.raise_for_status()

    return r.json()["response"]
```

## Exemplo: interpretação automática

```python
metricas = '''
RMSE = 0.42
MAE = 0.31
R² = 0.81
'''

prompt = f'''
Interprete cientificamente os seguintes resultados
de um modelo epidemiológico:

{metricas}
'''

print(interpretar(prompt))
```
