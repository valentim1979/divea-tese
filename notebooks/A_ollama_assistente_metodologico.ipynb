# A_ollama_assistente_metodologico

Integração do Ollama local ao ecossistema DIVEA.

## Objetivos
- auxílio metodológico
- revisão estatística
- geração de código R/Python
- interpretação epidemiológica

```python
import requests
import json
import pandas as pd
```

```python
OLLAMA_URL = "http://192.168.1.154:11434/api/generate"
MODEL = "divea-biostats:latest"

def perguntar_ollama(prompt, temperature=0.2):
    payload = {
        "model": MODEL,
        "prompt": prompt,
        "stream": False,
        "options": {
            "temperature": temperature
        }
    }

    response = requests.post(OLLAMA_URL, json=payload)
    response.raise_for_status()

    return response.json()["response"]
```

## Exemplo: revisão metodológica

```python
texto = '''
O estudo utiliza séries temporais epidemiológicas de SRAG
com modelos LSTM e TFT.
'''

prompt = f'''
Revise metodologicamente o seguinte delineamento:

{texto}
'''

print(perguntar_ollama(prompt))
```
